```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import plotly.express as px
```

# 3.8.3	Store the above data (the result of 3.8.2) in DataFrame and analyze districts that appear 3 or more than 3 years and print the corresponding state/ut, district, murders, and year in descending order.

```python
df = pd.read_csv('phase3(3.8.2).csv')
```

```python
df
```

```python
df.info()
```

```python
max_murder_district = df.groupby('district').filter(lambda x: len(x['year'].unique()) >= 3)
```

```python
max_murder_district
```

```python
max_murder_district['district'].value_counts()
```

```python
max_murder_districts = max_murder_district.sort_values(by='max_murder', ascending=False)
print(max_murder_districts[['state', 'district', 'max_murder', 'year']])
```

```python
max_murder_districts
```
